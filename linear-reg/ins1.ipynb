```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('insurance.csv')
df
```

```python
df.region.value_counts()
```

```python
le=LabelEncoder()
df.region=le.fit_transform(df.region)
df.sex=le.fit_transform(df.sex)
df.smoker=le.fit_transform(df.smoker)
df
```

```python
X=df.drop(['index','charges'],axis=1)
y=df.charges
X
```

```python
scaler_X=StandardScaler()
scaler_y=StandardScaler()
X_stan=scaler_X.fit_transform(X)
y_reshape=y.values.reshape(-1,1)
y_stan=scaler_y.fit_transform(y_reshape)
```

```python
reg=LinearRegression()
reg.fit(X_stan,y_stan)
```

```python
y_pred=reg.predict(X_stan)
y_pred_org=scaler_y.inverse_transform(y_pred)
```

```python
intercepts_st = reg.intercept_
coeff_st=reg.coef_
```

```python
std_y = scaler_y.scale_
mean_y = scaler_y.mean_
mean_X = scaler_X.mean_
```

```python
std_X = scaler_X.scale_
mean_X = scaler_X.mean_
std_y = scaler_y.scale_[0]  # Standard deviation for target
mean_y = scaler_y.mean_[0]  # Mean for target

# Destandardize coefficients
coefficients = coeff_st * (std_y / std_X)

# Destandardize intercept
intercept = intercepts_st * std_y + mean_y - np.sum(coefficients * mean_X)

print(f'Destandardized Coefficients: {coefficients}')
print(f'Destandardized Intercept: {intercept}')
```

```python
example_data = pd.DataFrame({
    'age': [19],
    'sex': [0],
    'bmi': [27.900],
    'children': [0],
    'smoker': [1],
    'region': [3]
})
prediction = scaler_X.transform(example_data)
pred_new=reg.predict(prediction)
predicted_s = scaler_y.inverse_transform(pred_new)
print(f'Predicted result ${predicted_s[0][0]:.2f}')
```
