```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('house_pred.csv')
df.head()
```

```python
df.info()
```

```python
X=df.drop(['date','price','street','city','country','statezip'],axis='columns')
pd.set_option('display.float_format', '{:,.2f}'.format)
X
```

```python
y=df.price
y
```

```python
scaler=StandardScaler()
X_stan=scaler.fit_transform(X)
X_stan
```

```python
#X_stan_df = pd.DataFrame(X_stan, columns=X.columns)
#X_stan_df
```

```python
#X_stan_df.describe()
```

```python
reg=LinearRegression()
reg.fit(X_stan,y)
```

```python
reg.intercept_,reg.coef_
```

```python
pd.set_option('display.float_format', '{:,.2f}'.format)
input_data = [[3.00, 2.00, 1930, 11947, 1.00, 0, 0, 4, 1930, 0, 1996, 0]]
input_data_standardized = scaler.transform(input_data)
pred = reg.predict(input_data_standardized )
print(pred)
```
