```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('Salary_dataset.csv')
```

```python
#X=data.drop(['index','Salary'],axis=1)
X=df[['Yearsofexp']]
y=df.Salary
```

```python
ax=sns.scatterplot(x='Yearsofexp',y='Salary',data=df)
ax.set(xlabel='Years of experience',ylabel='Salary',title='exp vs salary')
plt.show()
```

```python
#linear regression
reg=LinearRegression()
reg.fit(X,y)
```

```python
reg.intercept_,reg.coef_
```

```python
y_pred=reg.predict(X)
```

```python
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
```

```python
new_experience = np.array([[7]])
predicted_salary = reg.predict(new_experience)
print(f'Predicted salary for {new_experience[0][0]} years of experience: ${predicted_salary[0]:.2f}')
```

```python
#Standardization
scaler=StandardScaler()
X_st=scaler.fit_transform(X)
#y_reshaped = y.values.reshape(-1, 1)
#y_st = scaler.fit_transform(y_reshaped)
```

```python
plt.scatter(X_st[:, 0], y, color='blue', label='Age vs Charges')
plt.xlabel('Standardized Age')
plt.ylabel('Standardized Charges')
plt.title('Standardized Age vs Charges')
plt.legend()
```

```python
reg1=LinearRegression()
reg1.fit(X_st,y)
```

```python
reg1.intercept_,reg1.coef_
```

```python
y_pred1=reg1.predict(X_st)
plt.scatter(X_st, y, color='blue', label='Data Points')
plt.plot(X_st, y_pred1, color='red', linewidth=2, label='Regression Line')
```

```python
new_experienc = np.array([[7]])
predicted_salary = reg1.predict(new_experienc)
print(f'Predicted salary for {new_experienc[0][0]} years of experience: ${predicted_salary[0]:.2f}')
```
